```python
# load desired BankSim dataset into pandas
import pandas as pd
banksim_dataset_path = os.path.join(banksim_folder_path, "ieee_cis_joined_full.csv")
banksim_df = pd.read_csv(banksim_dataset_path)
```

```python
banksim_df = banksim_df.sort_values(by='TransactionDT', ascending=True)
cutoff_idx = int(len(banksim_df) * 0.70)
banksim_df = banksim_df.iloc[:cutoff_idx]
print(f"Subsetting complete: Using {len(banksim_df)} rows for EDA.")
```

```python
# preview BankSim dataset header
banksim_df.head()
```

```python
report = sv.analyze(banksim_df, pairwise_analysis="off")

# Save report locally in Colab
sweetviz_output_path = "/content/drive/MyDrive/SP2718F - GSSP2026/outputs/ieee_cis_report.html"

report.show_html(
    filepath=sweetviz_output_path,
    open_browser=False
)

print("Sweetviz report saved to:", sweetviz_output_path)
```

```python
# generate Sweetvis output and embed the saved HTML directly in Colab
from IPython.display import HTML, display

with open(sweetviz_output_path, "r", encoding="utf-8") as f:
    html_content = f.read()

display(HTML(html_content))
```
